```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

df[["ESA_tab_norm", "FCA_tab_norm"]] = scaler.fit_transform(
    df[["ESA_Score", "FCA_Score"]]
)
```

```python
df["ESA_text_norm"] = df["ESA_text"] / 2
df["FCA_text_norm"] = df["FCA_text"] / 2
```

```python
df["ESA_fused"] = 0.6 * df["ESA_tab_norm"] + 0.4 * df["ESA_text_norm"]
df["FCA_fused"] = 0.6 * df["FCA_tab_norm"] + 0.4 * df["FCA_text_norm"]
```

```python
df["Equity_Risk_Score"] = (
    0.5 * df["ESA_fused"] +
    0.5 * df["FCA_fused"]
)
```
